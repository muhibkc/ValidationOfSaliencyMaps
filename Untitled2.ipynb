```python
# get the data 
record_names = wfdb.get_record_list('but-pdb')
database = 'but-pdb'

#store data from different records
signals = []
p_anns = []
r_anns = []
samp_freq = []

for i, rec in enumerate(record_names):
    #load record
    record = wfdb.rdrecord(f"{int(rec):02}", pn_dir = database)
    if record.fs != 360:
        continue
    if i == 35:
        break
    

    #load signal and annotations
    sig = record.p_signal[:,0]
    p_ann = wfdb.rdann(f"{int(rec):02}", 'pwave', pn_dir=database)
    if len(p_ann.sample) < 10:
        continue
    r_ann = wfdb.rdann(f"{int(rec):02}", 'qrs', pn_dir = database)

    print(f"Record being collected is record number: {int(rec):02}")
    #store signals and annotations
    signals.append(sig)
    p_anns.append(p_ann)
    r_anns.append(r_ann)
    samp_freq.append(record.fs)
```

```python
#plot signal and P peaks
for i in range(5):
    sig_plot = signals[i][:360 * 5]
    t = np.arange(len(sig_plot)) / 360
    
    plt.plot(t, sig_plot)
    plt.title(f'First 5 seconds of signal from record {i+1}')
    plt.xlabel('Time (s)')
    plt.ylabel('Voltage (mV)')
    plt.show()
```

```python
# clean the signals
def cleanTheSignals(signals):
    clean_signals = []
    
    for i in range(len(signals)):
        sig_clean = signals[i].copy()
        sig_clean = sp.filter_X(sig_clean,360,band=[0.5], btype='highpass',ftype='SOS')
        sig_clean = sp.filter_X(sig_clean,360,band=[150], btype='lowpass',ftype='SOS')
    
        clean_signals.append(sig_clean)

    return clean_signals

clean_signals = cleanTheSignals(signals)
```

```python
plt.figure(figsize = (10,3))
plt.title(f' Plot of Clean vs Unclean Beat')

time = np.arange(len(signals[0][:p_anns[0].sample[1]])) / 360

plt.plot(time, signals[0][:p_anns[0].sample[1]], label = 'unclean')
plt.plot(time, clean_signals[0][:p_anns[0].sample[1]], label = 'clean')

plt.plot(p_anns[0].sample[:1] / 360,signals[0][p_anns[0].sample[:1]],'*')
plt.plot(p_anns[0].sample[:1]/360 ,clean_signals[0][p_anns[0].sample[:1]],'*')


plt.xlabel('Time (s)')
plt.ylabel('Voltage (mV)')
plt.legend()
plt.savefig('clean-v-uncleanbeat')
```
